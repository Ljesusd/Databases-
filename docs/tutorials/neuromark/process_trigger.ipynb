```python
import pathlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# filename = '/Users/jorge/Documents/Coding/Projects/personal-projects/motion-hub-toolbox/docs/tutorials/neuromark/emg/subject_07_cond_04_run_02_Trigger.csv'
filename = '/Users/jorge/Documents/Coding/Projects/personal-projects/motion-hub-toolbox/docs/tutorials/neuromark/emg/subject_07_cond_02_run_02_Trigger.csv'
trigger_fs = 2000

# Open trigger signal
df_trigger = pd.read_csv(filename)
trigger = df_trigger['Trigger']

# Look for the values to cut
split_start = np.where(trigger[0:int(len(trigger) / 2)] > 1)
split_end = np.where(trigger[int(len(trigger) / 2):] > 1)
start = split_start[0][0]
end = split_end[0][0] + int(len(trigger) / 2)

# Convert Start - End frames into time
start = start/trigger_fs
end = end/trigger_fs


plt.scatter( np.arange(0, len(trigger.values))/trigger_fs, trigger.values )
print(start)
print(end)

print(end-start)
```

```python
import pathlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# filename = '/Users/jorge/Documents/Coding/Projects/personal-projects/motion-hub-toolbox/docs/tutorials/neuromark/emg/subject_07_cond_04_run_02_Trigger.csv'
filename = '/Users/jorge/Documents/Coding/Projects/personal-projects/motion-hub-toolbox/docs/tutorials/neuromark/emg/subject_07_cond_02_run_02_Trigger.csv'
trigger_fs = 2000

# Open trigger signal
df_trigger = pd.read_csv(filename)
trigger = df_trigger['Trigger']

# Find the first and last peaks and plot them
first_peak = np.where(trigger > 1)[0][0]
last_peak = np.where(trigger > 1)[0][-1]

plt.plot(np.arange(0, len(trigger.values))/trigger_fs, trigger)

first_time = first_peak/trigger_fs
last_time = last_peak/trigger_fs
print(first_time)
print(last_time)

print(last_time-first_time)
```
