# convert_dir_c3d_to_eurobench:

This function converts a directory with c3d files to a Eurobench collection. The function uses the class C3DEurobench to convert each c3d file to a Eurobench object. The function returns a Eurobench collection with the points and analogs of each c3d file. It uses a regular expression (parameters `pattern_subject_condition_run`) to extract the subject, condition, and run from the c3d file name, being the pattern of the form `r"subject_(?P<subject>\w+)_cond_(?P<condition>\w+)_run_(?P<run>\w+)"`.
- `dict_filter_cols` is a dictionary with the keys 'Trajectories' and 'Analog'. The values are lists of strings with the columns to be filtered. If you want to filter all the columns starting with `*`, you can use the value `['!*']`. Here, `!` is used to exclude the columns. If you want to filter all the columns starting with `*`, except the USERMO columns, you can use the value `['!*', 'USERMO']`.
- You can also specify what to save using the parameters: `b_save_data` to save the trajectory data, `b_save_analogs` to save the analog data, and `dir_out` to specify the output directory. You can also save the events using the parameter `b_save_events` and the parameter `b_save_info` to save the information of the c3d file.

> **NOTE**  
> This converts to an eurobench collection

```python
import mh_toolbox.conversion.c3d.c3d_eurobench as eur
import importlib
import pathlib

importlib.reload(eur)
c3d_folder = pathlib.Path('../../../data/c3d/dataset1')
pattern = r"subject_(?P<subject>\w+)_cond_(?P<condition>\w+)_run_(?P<run>\w+)"
points_collection, analogs_collection = eur.convert_dir_c3d_to_eurobench(dir_in=c3d_folder,
                                                            b_save_data=False,
                                                            dict_filter_cols={'Trajectories': ['!*', '!USERMO']},
                                                            dir_out='',
                                                            pattern_subject_condition_run=pattern)
```
