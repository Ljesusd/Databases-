## extract_groups_from_regexp

```python
import mh_toolbox.base.process_filenames as pf
import re

pattern = r"subject_(?P<subject>\d+)_condition_(?P<condition>\w+)_run_(?P<run>\d+)_(?P<feature>\w+)"
match = re.search(pattern, "subject_1_condition_A_run_1_angles.csv")

process_file = pf.ProcessFilenames()
process_file._extract_groups_from_regexp(match, ['subject', 'condition', 'run', 'feature'])
vars(process_file)
```
