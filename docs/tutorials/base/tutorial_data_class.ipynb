```python
import unittest
from mh_toolbox.base.data_class import DataClass, InnerNest, OuterNest

class TestDataClass(unittest.TestCase):

    def setUp(self):
        self.data_class = DataClass(outer_nest_by='outer', inner_nest_by='inner')

    def test_add_outer_nest(self):
        outer_nest = OuterNest()
        self.data_class.add_outer_nest('test_outer', outer_nest)
        self.assertIn('test_outer', self.data_class.data_class)

    def test_add_nested_data(self):
        inner_nest = InnerNest()
        inner_nest.add_element(data=[1, 2, 3], time=[1, 2, 3], events=['event1', 'event2'], info={'info': 'test'})
        self.data_class.add_outer_nest('test_outer', OuterNest())
        self.data_class.add_nested_data('test_outer', 'test_inner', ['event1', 'event2'], [1, 2, 3], [1, 2, 3], {'info': 'test'})
        self.assertIn('test_inner', self.data_class.data_class['test_outer'].outer_nest)

    def test_has_data_data_class(self):
        outer_nest = OuterNest()
        inner_nest = InnerNest()
        inner_nest.add_element(data=[1, 2, 3], time=[1, 2, 3], events=['event1', 'event2'], info={'info': 'test'})
        outer_nest.add_inner_nest('test_inner', inner_nest)
        self.data_class.add_outer_nest('test_outer', outer_nest)
        self.assertTrue(self.data_class.has_data_data_class())

if __name__ == '__main__':
    unittest.main()
```
