# C3D file format
(https://medium.com/@yvanscher/explanation-of-the-c3d-file-format-c8e065300510)

The data probably has 2 sections, the 3D point data and the analog data.

- The 3D point data is 5D for each data point in the animation.
    - The first three values are usually x,y,z in 3D space.
    - The next two values are the error estimate (not sure what this means) and the number of cameras that registered this point (which should be a singular value, like 3, 4, 5,).

- The analog data can be anything that goes along with the 3D point data. For example it could be a measure of force or mean body temperature in each frame. This data often stored at a different rate than the 3D point data. The thing to remember is that this is not location data in 3D space.
- According to the creator of the format: Typical analog data in biomechanics contains force plate signals, EMG data, accelerometer data, and foot switch information for gait timing.
