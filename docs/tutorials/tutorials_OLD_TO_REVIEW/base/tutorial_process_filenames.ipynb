# Tutorial usage of process_filenames

## initialize_from_regexp

```python
import mh_toolbox.base.process_filenames as pf
filename = "subject_1_condition_A_run_1_angles.csv"
pattern = r"subject_(?P<subject>\d+)_condition_(?P<condition>\w+)_run_(?P<run>\d+)_(?P<feature>\w+)"
process_file = pf.ProcessFilenames()
process_file.initialize_from_regexp(filename, pattern)
vars(process_file)
```

## extract_groups_from_regexp

```python
import mh_toolbox.base.process_filenames as pf
import re

pattern = r"subject_(?P<subject>\d+)_condition_(?P<condition>\w+)_run_(?P<run>\d+)_(?P<feature>\w+)"
match = re.search(pattern, "subject_1_condition_A_run_1_angles.csv")

process_file = pf.ProcessFilenames()
process_file.extract_groups_from_regexp(match, ['subject', 'condition', 'run', 'feature'])
vars(process_file)
```

```python
import mh_toolbox.base.process_filenames as pf
import re

pattern = r"subject_(\w+)_condition_(\w+)_run_(\w+)_(\w+)"
match = re.search(pattern, "subject_1_condition_A_run_1_angles.csv")

process_file = pf.ProcessFilenames()
process_file.extract_groups_from_regexp(match, ['subject', 'condition', 'run', 'feature'])
vars(process_file)
```

## replace_capture_group_from_regexp

```python
import mh_toolbox.base.process_filenames as pf
import re

pattern = r"subject_(?P<subject>\w+)_condition_(?P<condition>\w+)_run_(?P<run>\w+)_(?P<feature>\w+)"
match = re.search(pattern, "subject_1_condition_A_run_1_angles.csv")

pf.replace_capture_group_from_regexp(pattern, 'subject', separator_prefix='_', value_to_replace='1')
```

## format_filename_from_regexp

```python
import mh_toolbox.base.process_filenames as pf
import re
filename = "subject_1_condition_A_run_1_angles.csv"
pattern = r"subject_(?P<subject>\w+)_condition_(?P<condition>\w+)_run_(?P<run>\w+)_(?P<feature>\w+)"
match = re.search(pattern, "subject_1_condition_A_run_1_angles.csv")

process_file = pf.ProcessFilenames(subject='1', condition='A', run='1', feature='angles')
process_file.format_filename_from_regexp(match, filename)

vars(process_file)
```
