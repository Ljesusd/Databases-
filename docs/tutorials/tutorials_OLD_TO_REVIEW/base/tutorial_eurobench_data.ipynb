## Transform data

### cleanunnamed_tolower_dataframe
This function is used to eliminate the unnamed columns and convert the columns to lower case. It also replaces double underscores by single underscores.

```python
import mh_toolbox.classes.eurobench_data as eb
import importlib
import pandas as pd
importlib.reload(eb)

# Create and clean dataframe
df = pd.DataFrame({'A__': [1, 2, 3], 'B': [4, 5, 6]})
eurobench_data = eb.EurobenchData()
eurobench_data.data = df 
eurobench_data.clean_unnamed_to_lower_dataframe()

print(eurobench_data.data)
```

### keep_drop_cols_dataframe
This function is used to filter the columns of a dataframe. It takes:
- `df`: Dataframe
- `list_filter_cols`: List of columns to keep or drop. If the column starts with `!`, it will be dropped. Otherwise, it will be kept. For instance, `['!Frame', 'Time']` will drop the column `Frame` and keep the columns `Time`

```python
import mh_toolbox.classes.eurobench_data as eb
import importlib
import pandas as pd
importlib.reload(eb)

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})
columns_to_filter_drop = ['!A']
eurobench_data = eb.EurobenchData(data=df)
eurobench_data.keep_drop_cols_dataframe(columns_to_filter_drop)
print(eurobench_data.data)
```

### curate_and_save_dataframes
This function is used to clean, rename and filter the columns of a dataframe. It uses the functions `cleanunnamed_tolower_dataframe` and `filter_cols_from_dataframe`. It takes:
- `df`: Dataframe
- `dict_rename_cols_data`: Dictionary with the columns to rename. The keys are the original column names and the values are the new column names
- `list_filter_cols_data`: List of columns to keep or drop. If the column starts with `!`, it will be dropped. Otherwise, it will be kept. For instance, `['!Frame', 'Time']` will drop the column `Frame` and keep the columns `Time`
- `output_file`: Path to the output file

```python
import mh_toolbox.classes.eurobench_data as eb
import importlib
import pathlib
import pandas as pd
importlib.reload(eb)

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})
dict_sets_data = {'file_ac': ['A', 'C'], 'file_b': ['B']}
dict_rename_cols = {'A': 'A_new', 'B': 'B_new', 'C': 'C_new'}
eurobench_data = eb.EurobenchData(data=df)
eurobench_data.curate_and_save_dataframes(dict_keep_drop_cols=dict_sets_data, 
                                     dict_rename_cols=dict_rename_cols, 
                                     dir_out=pathlib.Path.cwd(),
                                     filename_prefix='test'
                                     )
```
