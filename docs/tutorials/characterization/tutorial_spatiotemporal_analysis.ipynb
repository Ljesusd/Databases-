# SpatioTemporalAnalysisStrategy

```python
import numpy as np

from mh_toolbox.characterization.characterization import CharacterizationCollection
from mh_toolbox.characterization.spatiotemporal.spatio_temporal_features import SpatioTemporalAnalysisStrategy
import mh_toolbox.base.eurobench_element as eb
import mh_toolbox.base.eurobench_collection as ec

filename = '/Users/jorge/Documents/Coding/Projects/personal-projects/motion-hub-toolbox/docs/tutorials/neuromark/vicon/subject_07_cond_02_run_02_Trajectories.csv'
events = '/Users/jorge/Documents/Coding/Projects/personal-projects/motion-hub-toolbox/docs/tutorials/neuromark/vicon/subject_07_cond_02_run_02_gaitEvents.proc_carlota.yaml'
info = '/Users/jorge/Documents/Coding/Projects/personal-projects/motion-hub-toolbox/docs/tutorials/neuromark/vicon/subject_07_cond_02_run_02_Info.yaml'

element = eb.EurobenchElement(key_id='acceleration')
element.initialize(data_filename=filename, events_filename=events, info_filename=info)

split_events = element.split_element_using_events(events_category=('general_start_turn', 'general_end_turn'), events_key='events')

collection = ec.EurobenchCollection()
collection.add_dict_elements_to_collection(split_events)

type_characterization_col = CharacterizationCollection(SpatioTemporalAnalysisStrategy(), collection)
type_characterization_col.characterize_collection({'key_event': 'events', 'fs': 1000, 'plot': True})
```
